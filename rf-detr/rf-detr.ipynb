```python
from rfdetr import RFDETRMedium
from PIL import Image
import matplotlib.pyplot as plt
import os
```

```python
DATASET = "../dataset"
OUTPUT = "./outputs"
NUM_CLASSES = 6

EPOCHS = 50
BATCH_SIZE = 2
LEARNING_RATE = 1e-4
```

```python
model = RFDETRMedium(num_classes=NUM_CLASSES, pretrained=True)

model.train(
    dataset_dir=DATASET,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    lr=LEARNING_RATE,
    output_dir=OUTPUT
)
```

```python
def visualize_pred(image_path, conf=.3):
    image = Image.open(image_path)
    predictions = model.predict(image, conf=conf)

    masks = predictions.masks
    boxes = predictions.boxes
    class_ids = predictions.class_id
    scores = predictions.confidence

    plt.imshow(image)
    for mask in masks:
        plt.imshow(mask, alpha=.5)
    plt.axis('off')
```

```python
example_img = os.path.join(DATASET, "valid", "000013.jpg")
visualize_pred(example_img)
```
